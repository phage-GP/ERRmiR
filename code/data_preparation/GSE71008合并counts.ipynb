```python
import pandas as pd
import os
```

```python
count_path = '../../data/raw_counts/GSE71008/count_result/'
```

```python
files = os.listdir(count_path)      # 获取每个批次文件下文件名
```

```python
df = pd.read_csv(count_path+files[0],sep='\t',header=None,index_col=0,names=[files[0].replace('_count.txt','')])        # 每批以第一个文件为初始数据框，合并后续数据

for file in files[1:]:
    df1 = pd.read_csv(count_path+file,sep='\t',header=None,index_col=0,names=[file.replace('_count.txt','')])
    df = pd.concat([df,df1],axis=1)                   # 上游注释的文件miRNA顺序相同的
# df.to_csv(path+'合并.csv') # 每批生成一个文件
```

```python
df_expr = df.drop(['__no_feature','__ambiguous','__too_low_aQual','__not_aligned','__alignment_not_unique']) # 删除最后五行统计信息
```

一个样本对应两个SRR文件

```python
df_info = pd.read_csv('../../data/raw_counts/GSE71008/GSE71008_SraRunTable.txt')
```

```python
df_info.columns
```

```python
df_back = df_info[['Run','GEO_Accession (exp)','disease_type','sex']]
```

```python
df_comple = pd.merge(df_expr.T, df_back, left_index=True,right_on='Run',how='inner')
```
